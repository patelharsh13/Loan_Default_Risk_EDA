```python
def uplot(df, col, title, hue= None):
    
    sb.set_style('whitegrid')
    sb.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30
    
    temp = pd.Series(data = hue)
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 7 + 4*len(temp.unique())
    fig.set_size_inches(width,8)
    plt.xticks(rotation=45)
    plt.yscale('log')
    plt.title(title) 
    ax = sb.countplot(data = df, x = col, order = df[col].value_counts().index, hue = hue, palette='viridis')
    
    plt.show()
    
```

```python
uplot(targ_0,col = "NAME_INCOME_TYPE",title = "Income Type Distribution", hue = "CODE_GENDER")
```

# Correlation Analysis

```python
def targ_corr(data, title):
    plt.figure(figsize=(15, 10))
    plt.rcParams['axes.titlesize'] = 25
    plt.rcParams['axes.titlepad'] = 70
    
    sb.heatmap(data, cmap="RdYlGn", annot=False)
    
    plt.title(title)
    plt.yticks(rotation=0)
    plt.show()
```

```python
corr = targ_0.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
with sb.axes_style("white"):
    ax = sb.heatmap(corr, mask=mask, vmax=.3, square=True)
plt.title("Correlation Analysis of Target = 0")
```

Major Take-aways:

1. Age has higher negative correlation with amount credit of loan. That means young people has lower credit than older people.
2. Young people are more likely to pay off loans on time.
3. People who own an apartment has positive correlation with the amoung credit of loan.

```python
corr = targ_1.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
with sb.axes_style("white"):
    ax = sb.heatmap(corr, mask=mask, vmax=.3, square=True)

plt.title("Correlation Analysis of Target = 0")
```

Major Take-Away:
